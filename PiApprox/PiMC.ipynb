### Pi. MonteCarlo

Approximate Pi using montecarlo method 

```python
import numpy as np
from time import perf_counter
import matplotlib.pyplot as plt
```

```python
def estimate_pi(n=int(1e6)):
	"""Approximate PI - MonteCarlo method

	Parameters
	----------
	n : int, optional
		Number of points to use, by default int(1e5)

	Returns
	-------
	float
		Approximation of PI
	"""

	pts_in_circle = 0
	for _ in range(n):
		x = np.random.uniform(0,1)
		y = np.random.uniform(0,1)

		d = x**2 + y **2 

		if d <= 1:
			pts_in_circle += 1 

	return 4*pts_in_circle/n
```

```python
def piMC(n=int(1e5)):

	""" Approximate PI - MonteCarlo method
	
	Parameters
	----------
	n : int, optional
		Number of points to use, by default int(1e5)

	Returns
	-------
	pts_in : np.array
		Points inside the circle
	pts_out : np.array
		Points outside the circle

	"""
	
	pts_in = []
	pts_out = []
	for _ in range(n):
		x = np.random.uniform(0,1)
		y = np.random.uniform(0,1)

		d = x**2 + y **2 

		if d <= 1:
			pts_in.append([x,y])
		else:
			pts_out.append([x,y])

	return np.array(pts_in), np.array(pts_out) 
```

```python
def pltPI(pts_in, pts_out):
	"""Plot the points inside and outside the circle

	Parameters
	----------
	pts_in : np.array
		Points inside the circle
	pts_out : np.array
		Points outside the circle
	"""

	theta = np.linspace(0, np.pi/2, 100)
	x_circ = np.cos(theta)
	y_circ = np.sin(theta)

	plt.plot(pts_in[:,0], pts_in[:,1], 'rx', alpha=0.75, markersize=0.1)
	plt.plot(pts_out[:,0], pts_out[:,1], 'bx', alpha=0.5, markersize=0.1)
	plt.plot(x_circ, y_circ, 'k-', linewidth=0.1)
	plt.xlim(0, 1.1)
	plt.ylim(0, 1.1)
	plt.xlabel('x')
	plt.ylabel('y')
	plt.savefig('PiMC.png', dpi=1000)
	plt.show()
```

```python
pltPI(*piMC())
```
